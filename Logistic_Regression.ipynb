###BINOMIAL CLASSIFICATION

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, precision_score,recall_score,confusion_matrix

df = sns.load_dataset('diamonds')
price_s = df['price'].median()
df['df_prices'] = df['price'] > price_s
#encoding
df = pd.get_dummies(df,columns=['cut','color','clarity'],drop_first=True)
x = df.drop(columns=['price','df_prices'])
y = df['df_prices']
#training the data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model = LogisticRegression()
model.fit(x_train,y_train)
pred = model.predict(x_test)
#metric
print("accuracy",accuracy_score(y_test,pred))
print("precision",precision_score(y_test,pred))
print("recall",recall_score(y_test,pred))
print(classification_report(y_test,pred))
#visualisation
sns.heatmap(confusion_matrix(y_test,pred),cmap='coolwarm',annot=True,linewidths=1,linecolor='black')
plt.xlabel("actual value")
plt.ylabel("predicted value")
plt.title("binomial classification")
plt.show()
#adding column
df['pred']=model.predict(x)
df
```
